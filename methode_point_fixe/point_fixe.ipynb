# Implémentation du méhode point fixe en python

```python
import numpy as np
import matplotlib.pyplot as plt
from point_fixe import point_fixe
```

## La commande ci-dessous est la fonction que je souhaite approximée
Vous pouvez changée la valeur retournée de cette fonction mais il faut que cette fonction soit contractante c'est a dire $$|f'(x)|<1$$

```python
def f(x):
    return np.log(x**2 + 1)/2 + 3/2
```

```python
x0 = 1
x_steps = point_fixe(f,x0)

x_range = np.linspace(0,10,100)
y_range = f(x_range)

plt.figure(figsize=(10,5))
plt.plot(x_range, y_range, label='f(x)', color='red')
plt.axhline(0, color='black', linewidth=0.5)

for i in range(len(x_steps) - 1):
    x_curr = x_steps[i]
    x_next = x_steps[i+1]
    plt.plot(x_curr, f(x_curr), 'bo')
    plt.plot([x_curr,x_curr], [0, f(x_curr)], 'b--')
    plt.plot([x_curr,x_next], [f(x_curr), 0], 'g--')

plt.plot(x_steps[-1],f(x_steps[-1]),'go','solution approx')

plt.title('Methode du point fixe')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
