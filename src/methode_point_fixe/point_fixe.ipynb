# Implémentation du méhode point fixe en python

```python
import numpy as np
import matplotlib.pyplot as plt
from point_fixe import point_fixe
```

## La commande ci-dessous est la fonction que je souhaite approximée
Vous pouvez changée la valeur retournée de cette fonction mais il faut que cette fonction soit contractante c'est a dire $$|f'(x)|<1$$

```python
def g(x):
    return 2*x -np.log(x**2 + 1) + 3
def f(x):
    return np.log(x**2 + 1)/2 + 3/2
```

```python
x0 = 1
iter_pf = point_fixe(f,x0)

X = np.linspace(0,10,100)
Y = f(X)

plt.figure(figsize=(10,6))
#tracer la fonction
plt.plot(X, Y, label='f(x)', color='red', linewidth=2)
plt.axhline(0, color='black', linewidth=2)

# Tracer les iteration
for i in range(len(iter_pf) - 1):
    x_n = iter_pf[i]
    x_next = iter_pf[i+1]

    #Points
    plt.scatter(x_n, f(x_n), color="blue")

    #Segment verticale
    plt.plot([x_n,x_n], [0, f(x_n)], color="blue", linestyle="dashed")

    #Segment horizontale
    plt.plot([x_n, x_next], [f(x_n), f(x_n)], color="blue", linestyle="dashed")
    
plt.scatter(iter_pf[-1], f(iter_pf[-1]), color="blue", label="Itération point fixe")

plt.title('Méthode du point fixe')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
