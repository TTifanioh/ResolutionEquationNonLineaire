```python
import numpy as np 
import matplotlib.pyplot as plt 
from dichotomie import *
```

# Voici une fonction que je souhaite approximer la solution.
Vous pouvez changer cette fonction pour trouver la solution de votre fonction.

```python
def f(x):
    return x**2 -2
```

```python
a, b = 0, 2
points = dichotomie(f, a, b)

x_vals = np.linspace(0, 10, 400)
y_vals = f(x_vals)

# Creation du graphique
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="f(x) = sin(x)", color="blue", lw=2)
plt.axhline(0, color="black", lw=2) # Ligne de l'axe x

# Affichage des etapes
# for i,m in enumerate(points):
  #  plt.plot(m, f(m), "ro", alpha=(i+1)/len(points)) # point de plus en plus rouge
  #  plt.annotate(f"n={i}", (m, f(m)), textcoords="offset points", xytext=(0, 10), ha="center", fontsize=8)

plt.title("Convergence de la methode de dichotomie")
plt.xlabel("X")
plt.ylabel("f(X)")
plt.legend()
plt.grid(True)
plt.show()
```
