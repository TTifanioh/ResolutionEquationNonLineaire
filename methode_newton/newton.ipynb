# Implémentation d'algorithme de Newton-Raphson en python

```python
import numpy as np
import matplotlib.pyplot as plt
from newton import method_newton
```

## La commande ci-dessous est la fonction que je souhaite approximée la solution 
Vous pouvez changée la valeur retourner de f(x) ainsi que sa derivée df(x).

```python
def f(x):
    return 2*x - np.log(x**2 + 1) - 3 # fonction f(x)
def df(x):
    return 2 - (2*x/(x**2 + 1)) # derivée de f(x)
```

```python
x0 = 1.5
x_steps = method_newton(f,df,x0)

x_range = np.linspace(0,2,100)
y_range = f(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label="f(x)", color="blue")
plt.axhline(0, color="black", linewidth=0.5)

for i in range (len(x_steps) - 1):
    x_curr = x_steps[i]
    x_next = x_steps[i+1]
    plt.plot(x_curr, f(x_curr), "ro")
    plt.plot([x_curr, x_curr], [0, f(x_curr)], "r--")
    plt.plot([x_curr, x_next], [f(x_curr), 0], "g--")

plt.plot(x_steps[-1], f(x_steps[-1]), "go", "solution approx.")

plt.title("Methode de Newton")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid("True")
plt.show()
```

```python
def f1(x,y):
    return x**2 + y**2 - 2
def f2(x,y):
    return x**2 - y**2 - 1
```

```python
def newton_syst_lin(f1,f2,x0,y0,tol=1e-7,max_iter=100):
    x_vals = [x0]
    y_vals = [y0] 
    for i in range (max_iter):
        x = x_vals[-1]
        y = y_vals[-1]
        
        x_new = x - (3 - x*2)/(4*x)
        y_new = y - (1 - y*2)/(4*y)
        x_vals.append(x_new)
        y_vals.append(y_new)
        
        print(f"Etape {i+1} : x = {x_new:.6f}. f1(x) = {f1(x_new,y_new):.6f}")
        print(f"Etape {i+1} : y = {y_new:.6f}. f2(x) = {f2(x_new,y_new):.6f}")
        if abs(x_new - x) < tol :
        	break
        
    return x_vals,y_vals
```

```python
import numpy as np
import matplotlib.pyplot as plt
x0 = 1.5
y0 = 1
x_steps = newton_syst_lin(f1,f2,x0,y0)

x_range = np.linspace(0,2,100)
y_range = np.linspace(0,2,100)
y1_range = f1(x_range,y_range)
y2_range = f2(x_range,y_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y1_range, label="f1(x)", color="blue")
plt.plot(x_range, y2_range, label="f2(X)", color="red")
plt.axhline(0, color="black", linewidth=0.5)

for i in range (len(x_steps) - 1):
    x_curr = x_steps[i]
    y_curr = x_steps[i]
    x_next = x_steps[i+1]
    y_next = x_steps[i+1]
    plt.plot(x_curr, f1(x_curr), "ro")
    plt.plot(y_curr, f2(y_curr), "b*")
    plt.plot([x_curr, x_curr], [0, f1(x_curr)], "r--")
    plt.plot([x_curr, x_next], [f1(x_curr), 0], "g--")
    plt.plot([y_curr, y_curr], [0, f2(y_curr)], color="yellow", linestyle="v")
    plt.plot([y_curr, y_next], [f2(y_curr), 0], color="violet", linestyle="v")

plt.plot(x_steps[-1], f1(x_steps[-1]), "go", "solution approx.")
plt.plot(x_steps[-1], f2(x_steps[-1]), "bo", "solution approx.")


plt.title("Methode de Newton")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid("True")
plt.show()
```
