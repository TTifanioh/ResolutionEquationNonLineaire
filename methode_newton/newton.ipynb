# Implémentation d'algorithme de Newton-Raphson en python

```python
import numpy as np
import matplotlib.pyplot as plt
from newton import method_newton
```

## La commande ci-dessous est la fonction que je souhaite approximée la solution 
Vous pouvez changée la valeur retourner de f(x) ainsi que sa derivée df(x).

```python
def f(x):
    return 2*x - np.log(x**2 + 1) - 3 # fonction f(x)
def df(x):
    return 2 - (2*x/(x**2 + 1)) # derivée de f(x)
```

```python
x0 = 1
iter_newton = method_newton(f,df,x0)

X = np.linspace(0,10,100)
Y = f(X)

plt.figure(figsize=(10,6))
#Tracer de la fonction
plt.plot(X, Y, label='f(x)', color='red')
plt.axhline(0, color='black', linewidth=2)

#Tracer des iteration
for i in range(len(iter_newton) - 1):
    x_n = iter_newton[i]
    x_next = iter_newton[i+1]

    #Points
    plt.plot(x_n, f(x_n), color="blue")

    #Tagente
    slope = df(x_n)
    intercept = f(x_n) - slope*x_n
    X_tan = np.linspace(x_n-1, x_n+1, 50)
    Y_tan = slope * X_tan + intercept
    plt.plot(X_tan, Y_tan, color="blue", linestyle="dotted")

plt.scatter(iter_newton[-1], f(iter_newton[-1]), color="blue", label="Itération Newton")    


plt.title('Méthode de newton')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
